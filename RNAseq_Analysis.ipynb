```python
# Install fastp
!apt-get install -qq fastp

# Run fastp on our test sample
!fastp \
  -i fastq_files/SRR21820609_1.fastq \
  -I fastq_files/SRR21820609_2.fastq \
  -o fastq_files/SRR21820609_1_trimmed.fastq \
  -O fastq_files/SRR21820609_2_trimmed.fastq \
  -h fastqc_output/fastp_report.html \
  -j fastqc_output/fastp_report.json

# Check output
!ls -lh fastq_files/*trimmed*
```

```python
# Filter the GTF to only chr19 (using correct filename)
!grep "^chr19" reference/mm39_genes.gtf > reference/chr19_genes.gtf

# Verify it worked
!wc -l reference/chr19_genes.gtf
!head reference/chr19_genes.gtf
```

```python
# Install STAR aligner
!apt-get install -qq rna-star

# Verify installation
!STAR --version

# Build STAR index for chr19
# This will take ~5-10 minutes
!STAR --runMode genomeGenerate \
  --genomeDir star_index \
  --genomeFastaFiles reference/chr19.fa \
  --sjdbGTFfile reference/chr19_genes.gtf \
  --sjdbOverhang 99 \
  --genomeSAindexNbases 12 \
  --runThreadN 2

# Check what was created
!ls -lh star_index/
```

```python
# See what files were generated
!ls -lh star_index/
```

```python
# Run STAR alignment on our trimmed test sample
!STAR --genomeDir star_index \
  --readFilesIn fastq_files/SRR21820609_1_trimmed.fastq \
                fastq_files/SRR21820609_2_trimmed.fastq \
  --outFileNamePrefix alignment_output/sample_ \
  --outSAMtype BAM SortedByCoordinate \
  --runThreadN 2 \
  --quantMode GeneCounts

# Check what was created
!ls -lh alignment_output/
```

```python
# Check alignment quality
!cat alignment_output/sample_Log.final.out
```

```python
# First 20 lines of gene counts
!head -20 alignment_output/sample_ReadsPerGene.out.tab
```

```python
# Clean up and start fresh
!rm -rf star_index/*
!rm -rf reference/chr19*

# Download chromosome X
!wget -q -O reference/chrX.fa.gz \
  "http://hgdownload.soe.ucsc.edu/goldenPath/mm39/chromosomes/chrX.fa.gz"

# Uncompress
!gunzip reference/chrX.fa.gz

# Filter GTF to chrX genes
!grep "^chrX" reference/mm39_genes.gtf > reference/chrX_genes.gtf

# Check what we got
!ls -lh reference/
!wc -l reference/chrX_genes.gtf
```

```python
# Build STAR index for chrX
!STAR --runMode genomeGenerate \
  --genomeDir star_index \
  --genomeFastaFiles reference/chrX.fa \
  --sjdbGTFfile reference/chrX_genes.gtf \
  --sjdbOverhang 99 \
  --genomeSAindexNbases 13 \
  --runThreadN 2
```

```python
# Align to chrX
!STAR --genomeDir star_index \
  --readFilesIn fastq_files/SRR21820609_1_trimmed.fastq \
                fastq_files/SRR21820609_2_trimmed.fastq \
  --outFileNamePrefix alignment_output/sampleX_ \
  --outSAMtype BAM SortedByCoordinate \
  --runThreadN 2 \
  --quantMode GeneCounts

# Check results
!cat alignment_output/sampleX_Log.final.out | grep "Uniquely mapped"
!head -20 alignment_output/sampleX_ReadsPerGene.out.tab
```
